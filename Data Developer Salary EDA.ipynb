# Exploratory analysis questions and answers:
Q: Find total data developers jobs around the world for each year from 2020 to 2024?

```python
total_jobs_each_year_df = developers_salary_df.groupby('work_year').job_title.count()
```

```python
total_jobs_each_year_df
```

Q: Find 10 lowest salaries given to data developers?

```python
lowest_salaries_df = developers_salary_df.sort_values('salary_in_usd', ascending=True).head(10)
```

```python
lowest_salaries_df
```

Q: Find top 10 highest salaries given to data developers?

```python
highest_salaries_df = developers_salary_df.sort_values('salary_in_usd', ascending=False).head(10)
```

```python
highest_salaries_df
```

By finding the lowest and highest salaries we get to know that there aren't any outliers such as negative values which can be result of data entry errors, also we get to know that highest salaries are being paid in USA and by medium sized companies, however we will explore more about these later through visualization.

Q: Calculate average salaries of developers with respect to their job titles?

```python
avg_salaries_df = developers_salary_df.groupby('job_title').salary_in_usd.mean().sort_values(ascending=False).head(15)
```

```python
avg_salaries_df
```

Average salaries of data developers working in US companies

```python
developers_salary_df[developers_salary_df.company_location == 'US'].salary_in_usd.mean()
```

Q: Which job market is growing most in 2024 in field of data?

```python
jobs_2024_df = developers_salary_df[developers_salary_df.work_year == 2024]
```

```python
jobs_2024_df
```

```python
jobs_count_2024 = jobs_2024_df.groupby('job_title').salary_in_usd.count()
```

```python
jobs_count_2024.sort_values(ascending=False).head(1)
```

Data scientist is most growing job in 2024 around the world

## Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
developers_salary_df.job_title.unique().shape
```

```python
plt.figure(figsize=(12,9))
sns.barplot(avg_salaries_df);
plt.xticks(rotation=75);
plt.title('Average salaries for top 15 highest earning job titles')
```

```python
plt.figure(figsize=(9,6))
plt.title('Distribution of salaries based on experience level')
sns.scatterplot(x=developers_salary_df.work_year, y=developers_salary_df.salary_in_usd, hue=developers_salary_df.experience_level);
```

```python
large_company_df = developers_salary_df[developers_salary_df.company_size == 'L']
medium_company_df = developers_salary_df[developers_salary_df.company_size == 'M']
small_company_df = developers_salary_df[developers_salary_df.company_size == 'S']
```

```python
plt.figure(figsize=(9,6))
plt.title('Salaries of developers according to company size')
plt.hist([large_company_df.salary_in_usd, medium_company_df.salary_in_usd, small_company_df.salary_in_usd], bins=6, stacked=True)
plt.legend(['Large', 'Medium', 'Small']);
```

```python
plt.title('Total count of each type of employee in Data field')
sns.countplot(developers_salary_df.employment_type);
```
